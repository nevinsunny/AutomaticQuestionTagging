```python
import nltk
# Importing libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string
import re
import numpy as np
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()
 
# Cleaning the text sentences so that punctuation marks, stop words &amp; digits are removed
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    processed = re.sub(r"\d+","",normalized)
    y = processed.split()
    return y
```

```python
#Reading Datasets
train_clean_sentences = []
dataFrame=pd.read_csv('dataset/Fitness.csv')
x=dataFrame.values[:,0]
y1=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)
  
dataFrame=pd.read_csv('dataset/Sports.csv')
x=dataFrame.values[:,0]
y2=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Economics.csv')
x=dataFrame.values[:,0]
y3=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Electrical.csv')
x=dataFrame.values[:,0]
y4=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)


dataFrame=pd.read_csv('dataset/Diet-and-Nutrition.csv')
x=dataFrame.values[:,0]
y5=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Movie.csv')
x=dataFrame.values[:,0]
y6=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Dance.csv')
x=dataFrame.values[:,0]
y7=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Artificial-Intelligence.csv')
x=dataFrame.values[:,0]
y8=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Biology.csv')
x=dataFrame.values[:,0]
y9=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)



dataFrame=pd.read_csv('dataset/Chemistry.csv')
x=dataFrame.values[:,0]
y10=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Geo.csv')
x=dataFrame.values[:,0]
y11=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Astro_r.csv')
x=dataFrame.values[:,0]
y12=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Physics.csv')
x=dataFrame.values[:,0]
y13=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Biology_sp.csv')
x=dataFrame.values[:,4]
y14=dataFrame.values[:,0]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Chemistry_sp.csv')
x=dataFrame.values[:,0]
y15=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Geo_sp.csv')
x=dataFrame.values[:,0]
y16=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)

dataFrame=pd.read_csv('dataset/Economics_sp1.csv')
x=dataFrame.values[:,0]
y17=dataFrame.values[:,1]
for line in x:
  line = line.strip()
  cleaned = clean(line)
  cleaned = ' '.join(cleaned)
  train_clean_sentences.append(cleaned)
```
